# **Gerador de DATASET**

```python
import pandas as pd

# Exemplo de dados para treinamento
train_data = {
    'text': [
        "Este é um ótimo filme! Eu realmente gostei.",
        "O enredo é fraco e os personagens são chatos.",
        "Um dos piores filmes que já vi.",
        "Eu recomendaria este filme a todos os meus amigos."
    ],
    'label': [1, 0, 0, 1]  # 1 para positivo, 0 para negativo
}

# Exemplo de dados para avaliação/teste
test_data = {
    'text': [
        "Eu não poderia suportar assistir a este filme até o fim.",
        "A atuação dos atores é incrível!",
        "Houve alguns momentos engraçados, mas em geral, o filme foi decepcionante."
    ],
    'label': [0, 1, 0]  # 1 para positivo, 0 para negativo
}

# Criando DataFrames
train_df = pd.DataFrame(train_data)
test_df = pd.DataFrame(test_data)

# Salvando os DataFrames em arquivos CSV
train_df.to_csv("/content/drive/MyDrive/train.csv", index=False)
test_df.to_csv("/content/drive/MyDrive/test.csv", index=False)
```

# **Treinamento do modelo LLM**

```python
import pandas as pd
from simpletransformers.classification import ClassificationModel

# Carregando o conjunto de dados IMDb para classificação de sentimentos
train_data = pd.read_csv("/content/drive/MyDrive/train.csv")  # Substitua pelo caminho real do seu conjunto de treinamento
eval_data = pd.read_csv("/content/drive/MyDrive/test.csv")    # Substitua pelo caminho real do seu conjunto de teste

# Criando um modelo BERT para classificação binária
model = ClassificationModel(
    "bert",  # Pode ser substituído por outros modelos como "roberta", "distilbert", etc.
    "bert-base-uncased",  # Substitua pelo modelo pré-treinado desejado
    num_labels=2,  # Número de classes (positiva e negativa no caso de classificação de sentimento)
    args={"reprocess_input_data": True, "overwrite_output_dir": True},use_cuda=True
) # se usar uma GPU, o valor de use_cuda=True, senão False

# Treinando o modelo
model.train_model(train_data)

# Avaliando o desempenho do modelo no conjunto de teste
result, model_outputs, wrong_predictions = model.eval_model(eval_data)
```

# **Predição**

```python
# Realizando predições em novos exemplos
texts = ["Odiei este filme", "Odiei cada minuto deste filme."]
predictions, raw_outputs = model.predict(texts)

# Exibindo as predições
for text, prediction in zip(texts, predictions):
    print(f"Texto: {text}")
    print(f"Predição: {prediction}")
    print("=" * 30)
```
